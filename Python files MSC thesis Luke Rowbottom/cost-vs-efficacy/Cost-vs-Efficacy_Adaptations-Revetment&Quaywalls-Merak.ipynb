# Cost vs Efficacy of structural adaptations for revetments and quay walls (Case study: Port of Merak)

```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use("default")
```

## Revetments

```python
#Horizontal axis for adaptations, representing their efficacies
x = [0,0.5,1.0,1.5,2.0] # Full axis
x_berm = [0,0.4] #Axis for additional berm
x_parapet = [0,0.2] # Axis for parapet and crown wall
x_break1 = [0,0.5,1.0,1.3] #Data points low crested breakwater
x_break2 = [1.3,1.5,2.0] 

# Construction cost estimations for structural adaptations
lowcrested1 = [2283,2283,3239,4108] #Data points low crested breakwater
lowcrested2 = [4108,4759,6231] #low crested breakwater, unrealistic
Xbloc =[1190,1190,1725,2466,2997] # Construction costs Xblocs
Berm = [1104,1104] # Costs additional berm
Crown = [591,591] # Costs crown wall
Parapet = [487,487] # Costs parapet

plt.figure(figsize=(6,4.5))
plt.xlim(-0.1,2.1)
plt.grid()
plt.ylabel('Construction cost [/m]', size = 13)
plt.xlabel('Sea level rise [m]', size = 13)
plt.title('Construction cost structural adaptation vs sea level rise')
plt.plot(x_break1,lowcrested1,label = 'Low crested breakwater (LCB)', color = 'red',linewidth = 1.5, marker ='.', markersize = '8')
plt.plot(x_break2,lowcrested2, color = 'red',linewidth = 1.5, marker ='.', markersize = '8', linestyle = '--', alpha = 0.5)
plt.plot(x,Xbloc, label = 'Xblocs',color = 'blue',linewidth = 1.5, marker ='1', markersize = '12')
plt.plot(x_berm, Berm ,label = 'Additional berm', color = 'green',linewidth = 1.5, marker ='x', markersize = '10')
plt.plot(x_parapet,Crown, label = 'Crown wall', color = 'brown',linewidth = 1.5, marker ='^', markersize = '6')
plt.plot(x_parapet,Parapet, label = 'Parapet', color = 'black',linewidth = 1.5, marker ='*', markersize = '8')
plt.legend(bbox_to_anchor=(1.05, 1.005), loc='upper left', ncol = 1,  prop={"size":11})

#create table of construction costs within plot
columns = ('LCB', 'Xblocs', 'Berm', 'Crown wall', 'Parapet')
rows = [f"{0.5} m SLR" ,  f"{1.0} m SLR" , f"{1.5} m SLR", f"{2.0} m SLR"]

data = [[lowcrested1[1],Xbloc[1],'N/A','N/A', 'N/A'],
        [lowcrested1[2],Xbloc[2],'N/A','N/A','N/A'],
        [lowcrested2[1],Xbloc[3],'N/A','N/A','N/A'],
        [lowcrested2[2],Xbloc[4],'N/A','N/A','N/A'],]

plt.table(cellText=data, rowLabels= rows , loc = 'center', cellLoc='center', colLabels=columns, bbox=[0.0,-0.46,1,0.3])
props = dict(boxstyle='square', facecolor='grey', alpha=0.3)
```

## Quay walls

```python
#Horizontal axis for adaptations, representing their efficacies
x_short = [0,0.4] #Data points efficacy parapet
x_perf = [0,0.2] #Data points efficacy perforated quay wall
x_crest1 = [0,0.5,1.0,1.3] #Data points efficacy raising crest quay wall (realistic dimensions)
x_crest2 = [1.3,1.5,2.0] #Data points efficacy raising crest quay wall (unrealistic dimensions)
x_storm1 = [0,0.5,1.0,1.5] #Data points efficacy retractable storm wall (realistic dimensions)
x_storm2 = [1.5,2.0] #Data points efficacy retractable storm wall (unrealistic dimensions)

# Construction cost estimations for structural adaptations
crest1 = [17678,17678,30836,36099] #Data points construction costs raising crest quay wall (realistic dimensions)
crest2= [36099,41362,51889] #Data points construction costs raising crest quay wall (unrealistic dimensions)
Parapet = [245,245] #Data points constructin costs efficacy parapet
Storm1 = [798,798,1859,2769] #Data points construction costs retractable storm wall (realistic dimensions)
Storm2 = [2769,3830] #Data points construction costs retractable storm wall (unrealistic dimensions)
Perf = [2888,2888] #Data points construction costs perforated quay wall

plt.figure(figsize=(6,4.5))
plt.xlim(-0.1,2.1)
plt.grid()
plt.ylabel('Construction cost [/m]')
plt.xlabel('Sea level rise [m]')
plt.title('Construction cost structural adaptation vs sea level rise')
plt.plot(x_perf,Perf,label = 'Perforated quay wall', color = 'red',linewidth = 1.5, marker ='.', markersize = '8')
plt.plot(x_crest1,crest1, label = 'Raise crest',color = 'blue',linewidth = 1.5, marker ='*', markersize = '8')
plt.plot(x_crest2,crest2, color = 'blue',linewidth = 1.5, marker ='*', markersize = '8', linestyle = '--', alpha = 0.5)
plt.plot(x_storm1, Storm1 ,label = 'Retractable storm wall', color = 'green',linewidth = 1.5, marker ='x', markersize = '10')
plt.plot(x_storm2, Storm2 , color = 'green',linewidth = 1.5, marker ='x', markersize = '10', linestyle = '--', alpha = 0.5)
plt.plot(x_short,Parapet, label = 'Parapet', color = 'black',linewidth = 1.5, marker ='1', markersize = '12')
plt.legend(bbox_to_anchor=(1.05, 1.005), loc='upper left', ncol = 1,  prop={"size":11})

#create table of construction costs within plot
columns = ('Perforated wall', 'Raise crest', 'Storm wall', 'Parapet')
rows = [f"{0.5} m SLR" ,  f"{1.0} m SLR" , f"{1.5} m SLR", f"{2.0} m SLR"]

data = [['N/A',crest1[1],Storm1[1],'N/A'],
        ['N/A',crest1[2],Storm1[2],'N/A'],
        ['N/A',crest2[1],Storm2[0],'N/A'],
        ['N/A',crest2[2],Storm2[1],'N/A'],]     

plt.table(cellText=data, rowLabels=rows, loc = 'center', cellLoc='center', colLabels=columns, bbox=[0.0,-0.45,1,0.3])
props = dict(boxstyle='square', facecolor='grey', alpha=0.3)
```
